```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#fake data

wavelength_list = [500, 250, 670, 320, 300]
epoch_list = [1,2.5,4,5,6.7]
flux_matrix = np.empty((5,5))
flux_matrix.fill(2)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
X, Y = np.meshgrid(wavelength_list, epoch_list)
Z = flux_matrix

surf = ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap='hot')

plt.show()
```
